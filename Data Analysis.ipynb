## Quick Analysis of "Female hurricanes are deadlier than male hurricanes", Kiju Jung et al. (2014)
### Christopher Kotfila, July 6th 2014
 After having received some attention in the media, and seeing the scatter/analysis by @fnielsen [here](http://nbviewer.ipython.org/urls/gist.githubusercontent.com/fnielsen/ad93596c0e852b306ca5/raw/eb775510f5eb1afb49b40fe4591b48a7222652be/Nielsen2014Data_femalehurricanes.ipynb), I read the paper "Female hurricanes are deadlier than male hurricanes" by Kiju Jung published in PNAS.  Several of the data science bloggers and twitterers I followed were saying there were some issues with the analysis. In particular when you look at data after 1979 the effect described in the paper becomes insignificant - The author argues because the sample size is too small. When I saw that the supplemental data was available online,  I thought I would take a quick look for myself. This is as much an excuse to work on making matplotlib graphs as anything else.

 + doi: 10.1073/pnas.1402786111
 + The Paper can be found here: http://www.pnas.org/content/early/2014/05/29/1402786111
 + Supplimental data is available here: http://www.pnas.org/content/suppl/2014/05/30/1402786111.DCSupplemental
 


```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pylab as plt

import pandas as pd

DATA_FILE = "pnas.1402786111.sd01.xlsx"
mpl.rcParams["font.size"] = 18
```

```python
df = pd.read_excel(DATA_FILE, "Archival Study", skipfooter=6)
```

```python
ftnt = "$\dag$ Excluding Hurricane Katrina (1833 deaths, 2005) \n and Hurricane Audry (416 deaths, 1957) "
df["Gender"] = df.apply(lambda row: "Male" if not row["Gender_MF"] else "Female", axis=1)
ax = df.groupby("Gender").alldeaths.sum().plot(kind="bar", 
                                          grid=False, 
                                          color=["pink", "blue"], 
                                          figsize=(12., 8.), 
                                          ylim=(0, 1600),
                                          rot=0,
                                          title="All Deaths$^{\dag}$ by Gender of Hurricane Name \n 1950 - 2012")
ax.set_ylabel("All Deaths")
plt.figtext(0.12, 0.00,  ftnt , color='black', weight='roman', size='small')
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 8.))

df.groupby(["Category"]).alldeaths.sum().plot(kind="bar",
                                              ax=axs[0],
                                              grid=False, 
                                              color="purple", 
                                              ylim=(0, 600),
                                              title="All Deaths$^{\dag}$ by Hurricane Category \n 1950 - 2012")

axs[0].set_ylabel("All Deaths")

df.groupby("Category").size().plot(kind="bar",
                                   ax=axs[1],
                                   grid=False, 
                                   color=["green"], 
                                   ylim=(0, 50),
                                   rot=0,
                                   title="Counts$^{\dag}$ by Hurricane Category \n 1950 - 2012")
axs[1].set_ylabel("Counts")
plt.figtext(0.12, 0.00,  ftnt , color='black', weight='roman', size='small')
```

```python
title= "All Deaths$^{\dag}$ by Hurricane Category \n Breakdown by Gender (1950 - 2012)"
ax = df.groupby(["Category", "Gender_MF"]).alldeaths.sum().unstack("Gender_MF").plot(kind="bar",
                                                                                     legend=None,
                                                                                     stacked=True,
                                                                                     color=["blue", "pink"], 
                                                                                     grid=False,
                                                                                     ylim=(0, 600),
                                                                                     figsize = (12., 8.), 
                                                                                     title=title)
ax.set_ylabel("All Deaths")
handles, _ = ax.get_legend_handles_labels()
ax.legend(handles, ["Male", "Female"], title="Gender")
plt.figtext(0.12, 0.00,  ftnt , color='black', weight='roman', size='small')
```

```python
suptitle= "All Deaths$^{\dag}$ by Hurricane Category \n Breakdown by Gender with 1979 Split"

before_after = [("Hurricanes Before 1979", df[df['Year'] < 1979]),
                ("All Hurricanes", df),
                ("Hurricanes After 1978",  df[df['Year'] >= 1979])]

mpl.rcParams["font.size"] = 15

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(15., 5.), sharey=True)
fig.ylim = (0, 600)
for i,(title, tmp) in enumerate(before_after):
    axs[i].set_ylim(0, 600)
    tmp.groupby(["Category", "Gender_MF"]).alldeaths.sum().unstack("Gender_MF").plot(kind="bar",
                                                                                     legend=None,
                                                                                     ax=axs[i],
                                                                                     stacked=True,
                                                                                     color=["blue", "pink"],
                                                                                     title=title,
                                                                                     grid=False)
# Set the legend with better labels
axs[2].legend()
handles, _ = axs[2].get_legend_handles_labels()
axs[2].legend(handles, ["Male", "Female"], bbox_to_anchor=(1.2, 1.0))

axs[0].set_ylabel("All Deaths")
plt.figtext(0.12, -0.08,  ftnt , color='black', weight='roman', size='small')


fig.suptitle(suptitle, y=1.2, fontsize=18)
```

### What are the "Male" names pre-1979?

```python
df[(df["Year"] < 1979) & (df["Gender_MF"] == 0)][["Year", "Name", "Category", "alldeaths"]]
```

```python
import numpy as np
mpl.rcParams["font.size"] = 13

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16., 4.), sharey=True)


cmap = plt.cm.gist_earth
colors = cmap(np.linspace(0., 1., 6))

for i, (title, data_frame) in enumerate(before_after):
    for j, (cat, group) in enumerate(data_frame[['MasFem', 'alldeaths']].groupby(data_frame["Category"])):
        axs[i].scatter(group['MasFem'], group['alldeaths'], 
                       label="Category {}".format(cat), 
                       color=colors[j])
    axs[i].set_title(title)

axs[0].set_ylabel("All Deaths")
axs[2].legend(bbox_to_anchor=(1.6, 1.0))

axs[2].annotate( "Hurricane Sandy", 
                    xy=(9, 159), 
                    xytext=(1, 225), 
                    arrowprops=dict(facecolor='black', 
                                    width=2,
                                    shrink=0.1, 
                                    headwidth=6, 
                                    alpha=0.6))

plt.figtext(0.40, -0.06,  ftnt , color='black', weight='roman', size='small')

suptitle = "Scatter Plot for Male/Female Index by All Deaths$^{\dag}$ \n Breakdown by Category with 1979 split "
fig.suptitle(suptitle, y=1.2, fontsize=18)
```

### Comments
Unfortunately I don't have the specs for the author's negative binomial models, nor do I feel confident running my own regressions without really understanding the data - But looking at the scatter plot to the right,  it looks like hurricanes after 1979 (When the male/female alternating name practice started) are well balanced,  with the exception of a single outlier. In fact even with all the data (middle scatter plot) it appears to be well balanced with the exception of roughly 4 outliers.

The Post-1979 outlier of course was Hurricane Sandy,  Something I would not have guessed but is not surprising in retrospect is that Category 2 and 3 hurricanes appear to be more deadly - I would guess this is due to people not feeling like they should evacuate because it is not expected to be such a severe storm
